```python
def state_group(x):
    
    #nan values
    if not x==x:
        return x
    else:
        dct_ = {"Washington":"west_coast" ,
                "Arizona": "midwest" ,
                "Nevada": "west_coast" ,
                "California": "west_coast" ,
                "Oregon": "west_coast" ,
               }
        return dct_[x]


cs_2["state"] = cs_2["state"].apply(state_group)
```

# Plotting

## Plot: Number of responses

```python
# Number of responses:

number_of_response=cs_2["response"].value_counts()
number_of_response
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title("Number of responses")

sns.barplot(x=number_of_response.index ,y =number_of_response)
```

## Plot: Response rate by channel

```python
response_rate= cs_2.groupby(["sales channel",'response']).size().reset_index(name="Count")
```

```python
response_rate
```

```python
fig2, ax2 =plt.subplots ()

ax2.set_title("Response rate by channel")

sns.barplot(x=response_rate["response"], y=response_rate["Count"], hue= response_rate["sales channel"])
```

## Plot: response rate by the total claim amount

```python
response_total_claim=cs_2.groupby(["total claim amount", "response"]).size().reset_index()
```

```python
response_total_claim
```

```python
sns.catplot(x='response', y='total claim amount', data = cs_2, kind= "box")
```

## Plot: Response rate by income.

```python
response_income=cs_2.groupby(["response"]).count()["income"].reset_index()
```
